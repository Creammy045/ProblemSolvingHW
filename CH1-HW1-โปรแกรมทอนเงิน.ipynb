```python
# โปรแกรมการทอนเงินพร้อมระบบสต็อกเงิน

# 1) กำหนดหน่วยเงินเป็น "สตางค์"
denominations = [
    100000, 50000, 10000, 5000, 2000, 1000,   # 1000, 500, 100, 50, 20, 10 บาท
    500, 200, 100,                            # 5, 2, 1 บาท
    50, 25, 10, 5, 1                          # 50,25,10,5,1 สตางค์
]

print("=== โปรแกรมคำนวณเงินทอน ===")

try:
    # 2) รับอินพุตจากผู้ใช้ (บาท)
    price = float(input("ราคาสินค้า(บาท): "))
    paid = float(input("รับเงินมา(บาท): "))

    # 3) แปลงเป็นสตางค์
    price_st = round(price * 100)
    paid_st = round(paid * 100)

    # 4) ตรวจสอบว่าจ่ายพอหรือไม่
    if paid_st < price_st:
        print(f"รับเงินไม่พอ ต้องรับเพิ่มอีก {(price_st - paid_st)/100:.2f} บาท")

    else:
        change = paid_st - price_st
        print(f"\nเงินทอนทั้งหมด = {change/100:.2f} บาท\n")

        # 5) สต็อกเงินเริ่มต้น (แก้ได้ตามต้องการ)
        stock = {
            100000: 2,
            50000:  3,
            10000:  5,
            5000:   10,
            2000:   20,
            1000:   20,
            500:    10,
            200:    10,
            100:    20,
            50:     20,
            25:     20,
            10:     30,
            5:      30,
            1:      50,
        }

        print("=== รายการทอน ===")
        change_remaining = change
        change_used = {}

        # 6) คำนวณการทอนโดยใช้สต็อก
        for d in denominations:
            if change_remaining <= 0:
                break

            max_need = change_remaining // d
            available = stock[d]
            use = min(max_need, available)

            if use > 0:
                change_remaining -= use * d
                stock[d] -= use
                change_used[d] = use

                if d >= 100:
                    print(f"{d//100} บาท x {use}")
                else:
                    print(f"{d} สตางค์ x {use}")

        # 7) ตรวจสอบผลทอน
        if change_remaining > 0:
            print("\n สต็อกเงินไม่พอสำหรับการทอน!")
            print(f"ยังขาดเงินทอนอีก = {change_remaining/100:.2f} บาท")

        # 8) แสดงเฉพาะสต็อกคงเหลือปัจจุบัน
        print("=== สต็อกเงินคงเหลือปัจจุบัน ===")
        for d, amt in stock.items():
            if d >= 100:
                print(f"{d//100} บาท  : {amt} ชิ้น")
            else:
                print(f"{d} สตางค์ : {amt} ชิ้น")

except ValueError:
    print("กรุณากรอกจำนวนเป็นตัวเลขเท่านั้น")
```
